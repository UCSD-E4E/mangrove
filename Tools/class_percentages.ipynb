```
stats_pkl = "/content/drive/Shared drives/SIO and E4E Mangroves /Data/Classification_statistics.pkl"
stats_location = "/content/drive/Shared drives/SIO and E4E Mangroves /Data/Classification_statistics.xlsx"

class_stats = pd.read_pickle(stats_pkl)
class_stats
```

```
for index, row in tqdm(class_stats.iterrows()):
    full_path = row["full_path"]
    file = os.path.basename(full_path)
    folder = os.path.dirname(full_path)
    name = os.path.splitext(file)[0]
    dest_folder = os.path.join(folder,name+'_classifications/')

    mshp = join(dest_folder, 'm_' + name + '.shp')
    nmshp = join(dest_folder, 'nm_' + name + '.shp')

    m_gdf = geopandas.read_file(mshp)
    nm_gdf = geopandas.read_file(nmshp)

    m_gdf = m_gdf.to_crs({'init': 'epsg:3857'})
    nm_gdf = nm_gdf.to_crs({'init': 'epsg:3857'})
    
    class_stats.loc[index,"m_area"] = (m_gdf['geometry'].area/ 10**6).sum()
    class_stats.loc[index,"nm_area"] = (nm_gdf['geometry'].area/ 10**6).sum()

    class_stats.loc[index,"total_area"] = class_stats.loc[index,"m_area"] + class_stats.loc[index,"nm_area"]

    class_stats.loc[index,"m_percentage"] = class_stats.loc[index,"m_area"] / class_stats.loc[index,"total_area"]
    class_stats.loc[index,"nm_percentage"] = class_stats.loc[index,"nm_area"] / class_stats.loc[index,"total_area"]

    class_stats.loc[index, "last_updated"] = datetime.now()
```
